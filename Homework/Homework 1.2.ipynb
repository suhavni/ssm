## Homework 1.2 - Projectile motion under strong air resistance

Consider a projectile problem under gravity $\mathbf g$ and air resistance that is directly proportional to the square of its speed, i.e. $c |\mathbf v|^2$.

Given that $g = 9.8\;m/s^2$ and $c = 0.1\;m^{-1}$.

Subjected to the initial conditions:

$$x = 0, y = 0, v_x = 20 \; m/s \; \text{and}\;  v_y = 0$$

First **write down** the governing equation of motion.

Hence, **write down** the vector $\mathbf{v}$ and the matrix equation for $\frac{d\mathbf{v}}{dt}$.

**Solve** the problem for $x$ and $y$ on the time interval $0 \le t \le 10\; s$..

Then, **plot graph** of both components as function of time.

$
\frac{d\overrightarrow{v}}{dt} = \overrightarrow{g} - c|\overrightarrow{v}|^2 \hat{v}
$

$
|\overrightarrow{v}| = \frac{\overrightarrow{v}}{\hat{v}} \\
\hat{v} = \frac{\overrightarrow{v}}{|\overrightarrow{v}|} \\
$

$
\frac{d\overrightarrow{v}}{dt} = \overrightarrow{g} - c|\overrightarrow{v}|^2 \frac{\overrightarrow{v}}{|\overrightarrow{v}|} \\
\frac{d\overrightarrow{v}}{dt} = \overrightarrow{g} - c|\overrightarrow{v}| \overrightarrow{v}
$

$
  \overrightarrow{g} = -g \hat{j} \\
  \overrightarrow{v} = v_x \hat{i} + v_y \hat{j} \\
  |\overrightarrow{v}| = \sqrt{v_x^2 + v_y^2}
$

$
\frac{d\overrightarrow{v}}{dt} = -g \hat{j} - c\left(\sqrt{v_x^2 + v_y^2}\right) \left(v_x \hat{i} + v_y \hat{j}\right) \\
\frac{d\overrightarrow{v}}{dt} = -g \hat{j} - c v_x \sqrt{v_x^2 + v_y^2}\hat{i} - c v_y \sqrt{v_x^2 + v_y^2}\hat{j} \\
\frac{d\overrightarrow{v}}{dt} = - \left(c v_x \sqrt{v_x^2 + v_y^2}\right)\hat{i} - \left(g + c v_y \sqrt{v_x^2 + v_y^2}\right)\hat{j} \\
\frac{dv_x}{dt}\hat{i} + \frac{dv_y}{dt}\hat{j} = - \left(c v_x \sqrt{v_x^2 + v_y^2}\right)\hat{i} - \left(g + c v_y \sqrt{v_x^2 + v_y^2}\right)\hat{j}
$

$
\frac{dx}{dt} = v_x \\
\frac{dy}{dt} = v_y \\
\frac{dv_x}{dt} = - c v_x \sqrt{v_x^2 + v_y^2} \\
\frac{dv_y}{dt} = - g - c v_y \sqrt{v_x^2 + v_y^2}
$

$$
\frac{d\overrightarrow{v}}{dt} = - g - c  \left(v_x + v_y \right) \sqrt{v_x^2 + v_y^2}
$$

$$
\overrightarrow{v} = \left( {\begin{array}{*{20}{c}}
x\\
{{v_x}}\\
y\\
{{v_y}}
\end{array}} \right) = \left( {\begin{array}{*{20}{c}}
0\\
20\\
0\\
0
\end{array}} \right)$$

```python
from math import sqrt
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

g = 9.8
c = 0.1

        # v_x, v_y, x, y 
v_init = [20., 0, 0, 0]

t = np.arange(0, 10., 0.1)

def rhs_c(c, v, t):
    return [-c*v[0]*(sqrt(v[0]**2+v[1]**2)), -(g+c*v[1]*(sqrt(v[0]**2+v[1]**2))), v[0], v[1]]

v1 = odeint(lambda v, t: rhs_c(0, v, t) , v_init, t)

v2 = odeint(lambda v, t: rhs_c(c, v, t) , v_init, t)
```

```python
plt.figure(figsize=(10,7.5))
plt.plot(t, v2[:,2], lw=3)
plt.title("x-t graph", fontsize="xx-large", fontfamily="monospace")
plt.xlabel("t", fontstyle="oblique", fontsize="x-large")
plt.ylabel("x", fontstyle="oblique", fontsize="x-large")
plt.grid()
```
